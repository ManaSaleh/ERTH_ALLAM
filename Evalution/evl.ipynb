```python
df = pd.read_csv('answer_allam_v1.0.csv')
```

```python
df.info()
```

```python
from nltk.translate.bleu_score import sentence_bleu

df['BLEU_score'] = df.apply(lambda x: sentence_bleu([x['actual_answer']], x['ALLAM_answer']), axis=1)
```

```python
df['BLEU_score'].describe()
```

```python
from rouge import Rouge

rouge = Rouge()
# Calculate ROUGE scores and expand them into multiple columns (ROUGE-1, ROUGE-2, ROUGE-L)
df[['ROUGE-1', 'ROUGE-2', 'ROUGE-L']] = df.apply(lambda x: pd.Series(rouge.get_scores(x['ALLAM_answer'], x['actual_answer'])[0]), axis=1)
```

```python
df[['ROUGE-1', 'ROUGE-2', 'ROUGE-L']].describe()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['actual_answer'].tolist() + df['ALLAM_answer'].tolist())

cosine_similarities = [cosine_similarity(tfidf_matrix[i], tfidf_matrix[i + len(df)]) for i in range(len(df))]
df['Cosine_similarity'] = [sim[0][0] for sim in cosine_similarities]
```

```python
df['Cosine_similarity'].describe()
```

```python
from bert_score import score

P, R, F1 = score(df['ALLAM_answer'].tolist(), df['actual_answer'].tolist(), lang="ar")
df['BERTScore_Precision'] = P
df['BERTScore_Recall'] = R
df['BERTScore_F1'] = F1
```

```python
df[['BERTScore_Precision' , 'BERTScore_Recall' , 'BERTScore_F1']].describe()
```

```python
df['Exact_Match'] = df.apply(lambda x: int(x['actual_answer'] == x['ALLAM_answer']), axis=1)
```

```python
df['Exact_Match'].describe()
```
